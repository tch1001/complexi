```python
import complexi
import numpy as np
%matplotlib inline
plane = complexi.complexPlane('x')

def f(x):
    return x*x
```

```python
# plotting x (inputs)
increment = 0.05
for i in complexi.frange(0,1,increment): #imaginary part
    for j in complexi.frange(0,1,increment): #real part
        plane.drawLineColor(j+i*1j, j+increment+i*1j, 1, i,j,0.5)
```

```python
# plotting f(x)
increment = 0.05
for i in complexi.frange(0,1,increment): #imaginary part
    for j in complexi.frange(0,1,increment): #real part
        plane.drawLineColor(f(j+i*1j), f(j+increment+i*1j), 1, i,j,0.5)
```

```python
# plotting x (inputs)
magnitude = 3
for i in complexi.frange(0,np.pi, 0.05):
    plane.drawLine(0,magnitude*(np.cos(i)+np.sin(i)*1j), 10)
```

```python
# plotting f(x)
magnitude = 3
for i in complexi.frange(0,np.pi, 0.05): 
    plane.drawLine(0,f(magnitude*(np.cos(i)+np.sin(i)*1j)), 10)
```
